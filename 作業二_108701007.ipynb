# 全部月份 (由於新聞過少 從12-4月分析)

```python
import sys


def main(data_json):
    '''
    Read collected news as a dict objects list
    '''

    with open('/content/drive/MyDrive/chinatimes/2023041012_HW_2_begin.json', 'r') as read_file:
        lines = read_file.readlines()
        data_list = []
        for line in lines:
            data_list.append( eval(line)  )

        #print(data_list)
        print('Size of data_list:', len(data_list))
        print('View the first record:', data_list[0]['title'], data_list[0]['up_datetime'], data_list[0]['content'] )
        ####################################################
        # 同學們自行處理：資料讀出來之後，套用你選定的關鍵資訊演算法，並試著把分析結果視覺化
        from wordcloud import WordCloud
        import matplotlib.pyplot as plt
        import jieba
        import os
        
        # 將新聞標題轉為字串
        title_string = " ".join([data['title'] for data in data_list])
        
        # 設定字體
        font_path = 'KAIU.TTF'
        
        # 使用 jieba 套件進行中文斷詞
        cut_text = " ".join(jieba.cut(title_string))
        
        # 建立 WordCloud 物件
        wordcloud = WordCloud(width=800, height=800, background_color='white', font_path=font_path, min_font_size=10).generate(cut_text)

        # 繪製文字雲
        plt.figure(figsize=(8, 8), facecolor=None)
        plt.imshow(wordcloud)
        plt.axis("off")
        plt.tight_layout(pad=0)

        # 顯示文字雲
        plt.show()

        # 同學們自行處理：資料讀出來之後，套用你選定的關鍵資訊演算法，並試著把分析結果視覺化
        ####################################################

if __name__ == "__main__":
    if (len(sys.argv) < 2):
        print("Usage: %s <json file>"%sys.argv[0])
        sys.exit(1)

    if sys.argv[1]:
        main(sys.argv[1])
```
